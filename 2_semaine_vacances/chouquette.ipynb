```python
import requests
from lxml import html
import lxml
```

```python
#r = requests.get('http://lachouquette.ch/category/style/mode/')
categories = ['life','style','interviews','travel'];
tmpcat = ['life'];
```

```python
items = list()
```

```python
def extraction(lien, single):
    global items
    if(single):
        r = requests.get(lien)
        tree = html.fromstring(r.content)
        if(len(tree.xpath('//a[@class="author-name"]/text()'))):
            author = tree.xpath('//a[@class="author-name"]/text()')[0]
        else:
            author = ''
        if(len(tree.xpath('//li[@class="post-meta-date updated"]/text()'))):
            date = tree.xpath('//li[@class="post-meta-date updated"]/text()')[0]
        else:
            date = '';
        if(len(tree.xpath('//li[@class="post-meta-categories"]/a/text()'))):
            category = tree.xpath('//li[@class="post-meta-categories"]/a/text()')[0]
        else:
            category = ''
        if(len(tree.xpath('//h2[@id="comments-title"]/text()'))):
            comments = tree.xpath('//h2[@id="comments-title"]/text()')[0]
        else:
            comments = ''
        items = items + [[author, date, category, comments, lien]]
        print(author)
        print(lien)
    else:
        r = requests.get(lien)
        tree = html.fromstring(r.content)
        for link in tree.xpath('//h3[@class="title"]/a/@href'):
            extraction(link, True)
        if(len(tree.xpath('//a[@class="next page-numbers"]'))):
            extraction(tree.xpath('//a[@class="next page-numbers"]/@href')[0], False)
```

# Pandas integration

```python
import pandas as pd
import re
```

```python
df = pd.DataFrame(items)
df.columns = ['Author', 'Date', 'Category', 'Comments', 'Link']
```

```python
f = lambda x: re.findall("\t([a-zA-Zéäâ0-9û ]+)", x)
def g(x):
    if(len(x)):
        return x[0]
    else:
        return 'Sans'
```

```python
df["Author"] = df["Author"].apply(f).apply(g)
df["Date"] = df["Date"].apply(f).apply(g)
```

```python
def commentsApply(x):
    match1 = re.findall("([0-9]+) thoughts",x)
    if(len(match1)):
        return match1[0]
    match2 = re.findall("One thought",x)
    if(len(match2)):
        return 1;
    else:
        return 0
```

```python
df["Comments"] = df["Comments"].apply(commentsApply)
```

```python
df['Comments'] = pd.to_numeric(df['Comments'])
```
