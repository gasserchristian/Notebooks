# 4.1 Adding a 'weekday' column to our dataframe

```python
berri_bikes = broken_df[['Berri1']].copy()
```

```python
berri_bikes[:5]
```

    berri_bikes.index
    # are date objects
    dates = berri_bikes.index
    dates.weekday # day of the week
    dates.day # day coordinates

```python
berri_bikes.index.weekday
```

```python
berri_bikes.loc[:,'weekday'] = berri_bikes.index.weekday
```

Points importants :
* .loc prend argument ligne, colonne. Localisation
* [:,'weekday'] prend toutes les lignes et modifie ou crée la colonne weekday

# 4.2 Adding up the cyclists by weekday

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts
```

```python
weekday_counts.index = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekday_counts.plot(kind='bar')
```

```python
import numpy
weekday_counts = berri_bikes.groupby('weekday').aggregate(numpy.median)
weekday_counts.index = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekday_counts.plot(kind='bar')
```
