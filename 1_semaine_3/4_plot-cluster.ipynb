```python
import pandas as pd
import matplotlib
import numpy as np
import seaborn as sns
```

```python
cluster = pd.read_csv('../../data/cluster_dataset.csv')
```

```python
cluster.head()
```

```python
sns.pairplot(cluster[['Size','Relative Overlap','Time Coverage']],diag_kind='kde',plot_kws={'alpha':0.4,'s':3})
```

```python
cluster[:500000].plot.scatter(x='Relative Overlap',
                     y='Time Coverage',
                     s=cluster['Size'],alpha=0.2,figsize=(20,10))
```
